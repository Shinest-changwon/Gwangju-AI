##1.언어 분야의 제품 및 서비스, 기술 분석
* 제품 및 서비스
 
 * 트래비스사의 '터치 플러스'는 인공지능을 활용해 100개 이상의 언어를 실시간으로 번역 가능하며 두 가지 언어를 동시에 통역 가능  
 * 타임케틀사의 'WT2'는 전 세계 36개 언어를 지원하는 동시 통역이 가능한 무선 이어폰. 필요한 순간 버튼을 누를 필요없이 상대방의 음성이 인식되면 바로 번역이 되어 착용자의 이어폰에 번역된 음성이 전달
 
 * 국내 기업인 LG CNS가 공개한 'KorQuAD'는 1,550개의 Wikipedia Article에 대해 10,649건의 하위 문단과 크라우드 소싱을 통해 제작한 질의응답 등 한국어 표준데이터 10만 개 가량이 담겨 AI가 장문의 답변을 하는 데 활용될 전망

  * 구글이 발표한 AI 언어모델인 BERT(Bidirectional Encoder Representations from Transformers)는 자연어처리 평가인 SQuAD에서 인간(91)보다 높은 93점을 기록
 [이미지 - BERT: 구글의 AI 언어모델인 BERT는 자연어처리 평가인 SQuAD에서 인간(91점)보다 높은 93점을 기록했다. /SQuAD 홈페이지 캡처](https://image.chosun.com/sitedata/image/201902/08/2019020801816_1.png)

* 기술분석: 인간의 언어를 기계에서 처리하는 기술을 '자연어 처리'라고 하는데 자연어 처리의 근간을 이루는 언어 이해 개술은 '형태소 분석', '구문 분석', '개체명 분석', '화행 분석', '의도 분석'의 과정으로 이루어진다. 최근에는 딥러닝 방식을 활용해 언어 분야에서의 큰 발전을 이루고 있다. 대표적으로 개체명 분석 방면에서는 전통적으로 순서적 인식(Sequence Labeling)에서 좋은 성능을 발휘하는 CRF(Conditonal Random Field) 알고리즘을 많이 사용했으나 최근에는 딥러닝을 적용하여 개선한 LSTM(Long Short Term Memory)-CRF의 방법이나 LSTM-RNN(REcurrent Neural Network)을 널리 사용하고 있다.
 또한 문장에 나타난 긍, 부정 표현을 분석해 사용자의 의견을 정확히 파악할 수 있게 해주는 감성 분석이나 텍스트 형태의 비정형 데이터로부터 정형화된 정보를 추출하는 기술인 정보추출 분야에서도 딥러닝을 활용하고 있다.


#2. 음성 분야의 제품 및 서비스, 기술 분석

* 제품 및 서비스

 * 개인 비서 서비스: 애플의 '시리', 삼성전자의 '에스보이스', 엘지전자의 '큐보이스', 구글의 '구글나우', 마이크로소프트의 '코타나', 페이스북의 '엠'

 * 구글의 '구글 홈', SK텔레콤의 '누구', 네이버의 '클로버', 아마존의 '에코' 등 인공지능 스피커
    스마트 홈을 위한 사물 인터넷(IOT) 허브 역할 수행과 다양한 외부 서비스 활용(피자 주문, 렌터카 요청 등)
[이미지 - 인공지능 스피커](https://file.mk.co.kr/meet/neds/2017/11/image_readtop_2017_748209_15103800303096036.jpg)

 * 음성을 이용한 검색 및 텍스트 입력 기능


* 기술분석: 인간의 언어를 인공지능이 이해하기 위해서는 '언어 이해', '대화 관리', '응답 생성'의 과정이 필요하다. '언어 이해'에서는 텍스트 형태의 발화를 형태소, 개체명, 화행 및 의도 분석기능의 언어 분석기를 이용하여 사용자 의도를 파악하고, '대화 관리'에서는 적합한 대화 모델을 이용하여 응답을 위한 의미 구조를 생성하며, '응답 생성'에서는 추론과 수행 엔진을 이용하여 지식 베이스로부터 적합한 응답을 생성한다.
대표적으로 보면 대화 관리 과정에서는 '목적을 잘 수행하기 위해 어떤 대화 모델을 사용할 것인가'가 중요하다. 그 중 액티브 온톨로지(Active Ontology) 기반 대화 모델은 과업(Task)을 액티브 온톨로지 기반의 개념들로 나누어 정의하고, 언어 이해 결과와 추론 기반으로 대화 서비스를 진행하며, 수행 엔진 모듈을 이용해 외부 데이터 연계나 서비스를 수행한다. (애플의 '시리'개발에 사용)
또 다른 모델로는 구글 '알파고'와 같은 강화 학습을 사용하는 POMDP(Partially Observable Markov Decision Process) 기반 대화 모델이 있다. POMDP 기반 대화 모델은 대화 학습 말뭉치를 마코프 결정 프로세스를 통해 강화 학습한 대화 모델을 이용하여 최적의 응답을 생성하는 방법이다.

##3. 이미지 분야의 제품 및 서비스, 기술 분석 

 * 제품 및 서비스
  
  * DeepFake: 영화 부문에서 배우들의 실제 나이와 다른 나이대의 얼굴을 표현해 다양한 시대상을 보여줄 수 있다(영화 아이리쉬맨)

  * 스마트 팩토리: 전자제품에 들어가는 유리기판 제조 생산 공정 중 기포와 먼지를 구별하는 AI 이미지 분석 기술(LG CNS의 AI 이미지 분석 기술은 비전검사 영역에 적용되어 있음)

  * 보안 검색대: 기존에는 엑스레이로 사람이 하나하나 판별해야 했으나 AI 이미지 분석 기술 적용으로 보다 정확하게 분석해 자동으로 검색 가능

  * 무인 스토어: 구매자가 고른 제품을 계산대에 올리면 상품들을 자동으로 인식해 계산


 * 기술분석: DeepFake 기술을 적용한 영상으로 배우들의 다양한 나이대의 표정과 영화가 가지고 있는 캐릭터를 자동으로 분석해 출력할 수 있다. 기존의 그래픽 처리와 다른 점은 기존의 방식은 촬영 후 3D 모델링을 하고 작업자(사람)이 직접 하나하나 얼굴 표정과 보정 작업을 진행했다면 DeepFake기술은 AI가 데이터를 습득 후 사람의 손을 빌리지 않고 자동으로 얼굴을 바꿔줄 수 있다. 특히 2019년 작품 아이리쉬맨에서는 3명의 주연배우들의 모든 영화와 인터뷰 영상을 비롯한 샘플 영상(다양한 표정 촬영)을 Database화했으며, 단순히 배우들의 특정 나이대별 표정이 아닌 아이리쉬맨의 캐릭터에 맞는 표정을 표현했다는 점이 특별하다.

 스마트 팩토리 분별 공정, 보안 검색대, 무인 스토어 등 AI 이미지 분석 기술은 방대한 Data를 비롯해 머신러닝을 통해 AI가 스스로 정확한 이미지 판별을 내릴 수 있게 해준다. 그리고 이미지 분석을 넘어 이미지 자체를 생성하는 AI 기술도 큰 발전을 이루고 있다. GAN(Generative Adversarial Network)이라는 기술은 유사하지만 실제로 존재하지 않는 이미지를 생성할 수 있습니다.


#4. 자율주행 분야의 제품 및 서비스, 기술 분석

* 제품 및 서비스

 * 로보티즈의 '실외 자율주행로봇(배달로봇)': 2020년 4월부터 한 달간 서울 강서구 마곡 일대에서 점심식사 배달 (한 달 배달 건수: 685건)

  * 구호 활동 분야: 지진이나 해일 등 자연 재해를 당한 지역들의 생존자 위치 파악, 필요시 식료품 조달, 부상자 이송 등 감염과 공간적 제한이 있는 곳에 투입해 활동 가능

  * 물류 운송: 자율주행 이동우체국과 우편물 배달로봇, 집배원 추종로봇 등은 2020년 10월부터 시범 운용되며 구글의 자율주행 자회사 웨이모는 경쟁사인 피터빌트와 함께 자율주행 기술을 적용한 대형 트럭 개발 및 시범 운행 성공, 중국계 기업 Plus.ai(智加科技)는 레벨 4의 자율주행 트럭으로 버터 약 18톤을 적재하고 약 4,500키로미터를 달리는 데 성공.

    * 트위니의 '따르고': 사용자를 인식하는 카메라와 라이다, 초음파 센서를 탑재해 사용자가 움직이는 대로 따라다니는 자율주행 카트
     트위니의 '나르고': 여러 대 카트를 물리적으로 연결하지 않아도 앞에 있는 카트를 따라 이동하는 자율주행 카트(물류로봇)

* 기술분석: 자율주행은 레벨에 따라 구분을 하고 있는데 완전한 자율주행이 레벨5이며 운전자가 모든 것을 제어하는 것을 레벨0이라고 한다. 현재 수준은 레벨 2~3정도이며 레벨2는 고속도로에서 흔히 사용하는 '크루즈 컨트롤'이 있으며 이는 운전자의 주행을 보조하는 의미가 더 크다. 레벨3의 경우는 운전자가 항시 주시할 필요는 없지만 시스템에서 필요시 요청할 수 있는 수준이다. 레벨5의 경우에는 모든 도로구간에서 운전자가 없어도 자동차가 스스로 주행이 가능한 수준을 뜻한다.
자율주행 기술 분류에는 크게 차량과 인프라가 있다. 차량에는 인지, 판단, 제어 등이 있으며, 인프라에는 도로시설물, 노변센서, 교통센터, 통신 등이 있다. 그 중 차량부분을 살펴보면 인지에서는 주위 차량, 운전자, 도로, 장애물 등의 데이터를 수집해 주변환경을 인지하는 기술(GPS, 정밀지도, 라이더, 레이더, 카메라 등)이며 판단에서는 주행호나경에 따른 주행상황을 인식하고 최적의 주행조건(경로, 속도 등)을 결정하는 기술(주행경로 탐색, 차량/보행자 충돌방지, 장애물 회피 등)이며 제어는 차량 주행 및 움직임과 관련된 구동계 등을 제어하는 기술(종방향, 횡방향 제어)이다. 자율주행에서 AI의 역할은 인프라와 인지 기술을 통해 주고 받은 Data를 통해 최적의 판단을 도출해 구동계를 제어할 명령을 내리는 것이다.
